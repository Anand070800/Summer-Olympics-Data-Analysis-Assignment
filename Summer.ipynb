### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
cities = []
i=1

for city in df['City'].unique():
    cities.append(city)
print(f"There are {len(cities)} cities in Summer Olympics")
for j in cities:
    print(f"{i} {j}")
    i+=1
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
df.head()
sport=[]
for Sport in df['Sport'].unique():
    sport.append([Sport,len(df.loc[(df['Sport'] == Sport) & (df['Medal'] == 'Gold')])])
data=pd.DataFrame(sport,columns = ['Sport','Gold_Medal']).sort_values(by='Gold_Medal', ascending = False)   
print(data.head())
my_colors =['b', 'r', 'g', 'y', 'k']
data.head().plot(x = 'Sport', y = 'Gold_Medal', kind = 'bar', figsize = (12,5),stacked=True,color=my_colors)
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
new_df=df.loc[(df['Country'] == 'IND')&(df['Medal'] == 'Gold')]
new_df1=new_df[['Year','Country','Medal']]
#print(new_df1.sort_values(by='Year', ascending = True))
print("According to above data india won first Medal in 1928 ")
year_gold=[]
for Year in new_df1['Year'].unique():
    year_gold.append([Year,len(df[df['Year'] == Year])])
data=pd.DataFrame(year_gold,columns = ['Year','Num_Medal']).sort_values(by='Year', ascending = True)   
print(data.head())
my_colors =['b', 'r', 'g', 'y', 'k']
data.head().plot(x = 'Year', y = 'Num_Medal', kind = 'bar', figsize = (12,5),stacked=True,color=my_colors)
```

### 8. Which sport is having most female Gold Medalists? (Top 5)
